```python
from pyspark import SparkConf, SparkContext
```

```python
conf = SparkConf().setMaster("local").setAppName("MostPopularMovie")
sc = SparkContext(conf = conf)
```

```python
lines = sc.textFile("file:///home/turan/SparkCourse/section_4/ml-100k/u.data")
```

```python
parsedLines = lines.map(lambda x: x.split('\t')[1])
```

```python
movieTuples = parsedLines.map(lambda x: (x, 1))
```

```python
movieCountsRDD = movieTuples.reduceByKey(lambda x, y: x + y)
```

```python
sortedMoviesRDD = movieCountsRDD.map(lambda x: (x[1], x[0])).sortByKey(ascending=False)
```

```python
results = sortedMoviesRDD.collect()
```

```python
for count, movie in results:
    print(f"Film ID: {movie}, Sayı: {count}")
```
